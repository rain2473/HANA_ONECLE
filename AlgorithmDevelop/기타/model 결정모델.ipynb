```python
from DB import DataFrameHandler
from DB import AIModelHandler
import pandas as pd
import pickle, time
pd.options.mode.chained_assignment = None

ohlcv = []
fundamental = []
```

```python
total_time = 0
for isin in isins:
    start_time = time.time()
    tmp = DataFrameHandler.findItems('tmp_ohlcv', condition=f"isin = '{isin}'")
    end_time = time.time()
    elapsed_time = end_time - start_time

    print(f"{isin} 종목 쿼리 실행 소요 시간: {elapsed_time:.4f} 초")
    total_time += elapsed_time
    ohlcv.append(tmp)
print(f"{len(ohlcv)}건 쿼리 실행 소요 시간: {total_time:.4f} 초")
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns

def calcFeatureImportance(X, y, n_estimators=100, random_state=29):
    rf_model = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state)
    rf_model.fit(X, y)
    feature_importances = rf_model.feature_importances_
    feature_importance_df = pd.DataFrame(feature_importances).T
    feature_importance_df.columns = X.columns
    return feature_importance_df

def preprocessing(i):
    tmp1 = ohlcv[i].sort_values("S_DATE").reset_index(drop=True)
    tmp2 = fundamental[i].sort_values("S_DATE").reset_index(drop=True)
    tmp = pd.merge(tmp1,tmp2, how = 'outer')
    tmp.drop(columns=['ISIN', 'AMOUNT'], inplace=True)
    tmp['S_DATE'] = tmp.index
    y1 = pd.DataFrame(tmp['CLOSE']).rename(columns={'CLOSE':'target1'})
    y2 = pd.DataFrame(tmp['UPDOWN']).rename(columns={'UPDOWN':'target2'})
    y2 = (y2 > 0) * 1
    y1 = y1.drop(y1.index[0]).reset_index(drop=True)
    y2 = y2.drop(y2.index[0]).reset_index(drop=True)
    tmp = tmp.drop(tmp.index[-1]).reset_index(drop=True)
    tmp.fillna(0, inplace=True)
    return1 =  pd.concat([tmp, y1, y2], axis=1)

    return2 = calcFeatureImportance(tmp, y1['target1'])
    return3  = calcFeatureImportance(tmp, y2['target2'])
    return return1, return2, return3

with open('ohlcv.sav', 'rb') as file:
    ohlcv = file.read()
ohlcv = pickle.loads(ohlcv)
with open('fundamental.sav', 'rb') as file:
    fundamental = file.read()
fundamental = pickle.loads(fundamental)
```

```python
def show_summary_statistics(dataframes, labels):
    num_plots = len(dataframes)
    top_n = 13

    fig, axes = plt.subplots(num_plots, figsize=(10, 6 * num_plots))

    for i, (df, label) in enumerate(zip(dataframes, labels)):
        ax = axes[i] if num_plots > 1 else axes

        df = df.loc[['mean', 'std', '50%']]

        sns.barplot(x=df.columns, y=df.loc['mean'], ax=ax)
        ax.set_title(f'통계량 요약 - {label}')
        ax.set_ylabel('값')
    
    plt.tight_layout()
    plt.show()
```

```python
show_summary_statistics(data_to_plot, labels)
```
