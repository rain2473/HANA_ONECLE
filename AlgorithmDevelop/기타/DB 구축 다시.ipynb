```python
from pykrx import stock
from DB import DataFrameHandler
from DB import AIModelHandler
import pandas as pd
import pickle

pd.options.mode.chained_assignment = None
```

```python
tickers = AIModelHandler.selectAllStock()
tickers
```

```python
failed = []
for i, ticker in enumerate(tickers):
    try:
        ohlcv = stock.get_market_ohlcv("20130101", "20230831", ticker)
        fundamental = stock.get_market_fundamental("20130101", "20230831", ticker)
        ohlcv['ISIN'] = str(ticker)
        fundamental['ISIN'] = str(ticker)
        ohlcv = ohlcv.reset_index().rename(columns={'날짜':'S_DATE', '시가':'OPEN', '고가':'HIGH', '저가':'LOW', '종가':'CLOSE', '거래량':'VOLUME', '거래대금':'AMOUNT', '등락률':'UPDOWN'})
        fundamental = fundamental.reset_index().rename(columns={'날짜':'S_DATE'})
        DataFrameHandler.insertItems('tmp_ohlcv', ohlcv)
        DataFrameHandler.insertItems('tmp_fundamental', fundamental)
        print(f"{ticker} insert 성공 : {i + 1}개 완료")
    except Exception as e:
        print(e)
        failed.append(ticker)
pickle.dump(failed, open('failed.sav', 'wb'))
```
